```python
def winsorize(name):
    ''' Winsorize abnormal data
    '''
    df = train[name].copy()
    mean = df.mean()
    std = df.std()
    ind = (df > mean + 3 * std)
    df[ind] = mean + 3 * std
    ind = (df < mean - 3 * std)
    df[ind] = mean - 3 * std
    return df
```

```python
def step_error(x, y, group = 5, num = True):
    
    df = pd.concat( [x, y], axis = 1 ).fillna(0)
    df.columns = ['x', 'y']
    
    df_ = pd.DataFrame( df.sort_values(by = 'x') )
    if num:
        step = df_.shape[0] / group
        df_['group'] = np.arange( df_.shape[0] ) // step
    else:
        df_['group'] = df['x']
    return df_.groupby('group').apply(lambda x: pd.Series( [x['x'].mean(), x['y'].mean(), x['y'].std()] ) )
    
```

```python
def my_plot(ax, name, group = 5, num = True, y_name = 'SalePrice'):
    x = train[ name ].copy()
    y = np.log( train[y_name] )
    is_str = any( [type(i) == str for i in x.unique()] )
    if num:
        x = x.fillna(-1.)
        
    else:
        if is_str:
            x = x.fillna('NaN')
            x = train[name]
            unique_values = x.unique()[ x.unique() == x.unique() ]
            dict_ = dict( zip( unique_values, np.arange(unique_values.shape[0]) + 1 ))
            if any( x.unique() != x.unique() ) :
                dict_['NaN'] = -1.
            x = x.map( dict_ ).fillna(-1.)
            name_series = pd.Series(dict_).sort_values()
        else:
            
            unique_values = x.unique()[ x.unique() == x.unique() ]
            dict_ = dict( zip( unique_values, unique_values ) )
            if any( x.unique() != x.unique() ) :
                dict_[-1.] = -1.
            x = x.fillna(-1.)
            name_series = pd.Series(dict_).sort_values()
        
        
    ax.plot(x, y, '*', alpha = 0.01, label = name)
    ax.legend(loc = 'best')
    error = step_error(x, y, group = group, num = num)
    ax.errorbar(error[0], error[1], error[2], uplims=True, lolims=True,)
    if (not num):
        plt.xticks( name_series.values )
        plt.gca().set_xticklabels(name_series.index) 
    
    return error
```

```python
def my_difference(name):
    # count amount of *casual* of different *hour* groups
    count_casual = train.groupby(name).apply(lambda x: x['SalePrice'].mean())

    # find the peak hour
    peak = count_casual[count_casual == count_casual.max()].index.values[0]
    print('Peak %s is %.0f.'%(name, peak) )

    # plot *casual* amounts of different *hours*
    count_casual.plot(grid = True)
    plt.show()

    # calculate hour difference
    diff = train[name] - peak
    while abs( abs( diff.max() ) \
          - abs(diff.min() ) ) >= 2:
        if abs( diff.max() ) > abs(diff.min() ):
            ind = diff.values == diff.max()
            diff.values[ind] = diff.min() - 1
        else:
            ind = diff.values == diff.min()
            diff.values[ind] = diff.max() + 1
    return abs( diff )
```

```python
def plot_columns(selected_columns = selected_columns, label_lst = label_lst):
    n_row = int( np.ceil( len(selected_columns) / 4 ) )
    fig = plt.figure(figsize = [15, 3 * n_row])

    for i, name in zip(np.arange(len(selected_columns)), selected_columns):
        ax = fig.add_subplot( n_row, 4, int(i + 1) )
        if name in label_lst:
            error = my_plot(ax, name = name, num = False)

        else:

            error = my_plot(ax, name = name, group = 20)
    plt.show()
    return
```
